```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(x = iris.petal_width , y = iris.petal_length)
```

```python
sns.displot(iris['sepal_length'])
```

```python
tips = sns.load_dataset("tips")
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x =tips.total_bill , y = tips.tip , data = tips , style = 'size',hue = "time" )
```

```python
sns.catplot(x = 'day' , y = 'total_bill' , data  = tips)
```

```python
sns.jointplot(x = tips.total_bill , y = tips.tip)
```

```python
sns.pairplot(tips)
```

```python
# pair plot
sns.pairplot(iris)
```

```python
# line plot
sns.lineplot(data = tips , x = 'size' , y = 'total_bill')
plt.title('Line Plot of Total Bill vs Size ')
```

```python
# box plot 
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# violoin plot
sns.violinplot(data=tips,x="size",y="total_bill")
```

```python
# histograms
sns.histplot(tips['total_bill'], bins=10, kde=True)
```

```python
# KDE Plot
sns.kdeplot(tips["total_bill"], shade=True)
```

```python
# Heat Map
sns.heatmap(tips[['total_bill', 'tip', 'size']].corr(), annot = True, cmap="coolwarm")
```
