### indexing in python

* Dataframe.[ ] ; This function also known as indexing operator
* Dataframe.loc[ ] : This function is used for labels.
* Dataframe.iloc[ ] : This function is used for positions or integer based
* Dataframe.ix[] : This function is used for both label and integer based

```python
data = pd.read_csv("nba.csv", index_col ="Name")

#  indexing operator
first = data["Age"]
first = data[["Age", "College", "Salary"]]

### loc operator
first = data.loc["Avery Bradley"]
rows = data.loc[["Avery Bradley", "R.J. Hunter"]]

# retrieving two rows and three columns by loc method
first = data.loc[["Avery Bradley", "R.J. Hunter"],["Team", "Number", "Position"]]
# Extracting rows between two index labels
rows = data.loc["Avery Bradley":"Isaiah Thomas"]

# retrieving all rows and some columns by loc method
first = data.loc[:, ["Team", "Number", "Position"]]

### iloc operator
row2 = data.iloc[3]

# retrieving multiple rows by iloc method
row2 = data.iloc [[3, 5, 7]]

# retrieving two rows and two columns by iloc method
row2 = data.iloc [[3, 4], [1, 2]]

# retrieving all rows and some columns by iloc method
row2 = data.iloc [:, [1, 2]]

### ix operator
first = data.ix[1]
first = data.ix["Avery Bradley"]
x2 = data.ix[:4, 1:4]
x1 = data.ix[10:20, 'Height']
```

### Pandas Series.str.slice()

str.slice() method is used to slice substrings from a string present in Pandas series object.
Syntax: Series.str.slice(start=None, stop=None, step=None)

```python
# slicing till 2nd last element
data["Salary (int)"]= data["Salary"].str.slice(start=1, stop=-1, step=1)
```

```python
## column-slices of DataFrame

# Method 1: Slice Columns in pandas using reindex
df2 = df1.reindex(columns = ['c','b'])

# Method 2: Slice Columns in pandas using loc[]
# Syntax: [ : , first : last : step]
df2 = df1.loc[:, "b":"d":2]

# Method 3: Slice Columns in pandas using iloc[]
df2 = df1.iloc[:, 1:3:1]
```
