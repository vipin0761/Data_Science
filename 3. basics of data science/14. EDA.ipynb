```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
wine_quality = fetch_ucirepo(id=186) 

# data (as pandas dataframes) 
X = wine_quality.data.features 
Y = wine_quality.data.targets 
df = pd.concat([X,Y],axis=1)
df.head()
```

```python
## summary of the dataset 
df.info()
```

```python
## descritive summary of the dataset
df.describe()
```

```python
# column analysis
df.columns
df['quality'].unique()
```

```python
df['quality'].value_counts()
```

```python
# missing values
df.isnull().sum()
```

```python
# check for duplicated records
df[df.duplicated()].head()
```

```python
# removing duplicated records
df.drop_duplicates(inplace=True)
```

```python
# Correlation analysis
df.corr()
```

```python
# correlation heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

```python
df.quality.value_counts().plot(kind='bar')
```

```python
sns.histplot(df['fixed_acidity'], kde=True)
```

```python
# categorical plot
sns.catplot(x="quality",y="alcohol",data=df, kind='box')
```

```python
# segregate numerical and categorical features in a dataset
numerical_features = [feature for feature in df.columns if df[feature].dtype!='O']
categorical_features = [feature for feature in df.columns if df[feature].dtype=='O']
```

```python
## visualization diagrams
from altair import FontWeight
from pyparsing import alphanums


plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis on Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(len(numerical_features)):
    plt.subplot(5,4,i+1)
    sns.kdeplot(x=df[numerical_features[i]],shade=True, color='red')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

```python
# univariate, bivariate and multivariate analysis
sns.pairplot(df)


# categorical columns
plt.Figure(figsize=(20,15))
plt.suptitle('Univariate Analysis on Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
for i in range(len(categorical_features)):
    plt.Subplot(5,4,i+1)
    sns.countplot(x=df[categorical_features[i]],palette='Set2')
    plt.xlabel(categorical_features[i])
    plt.xticks(rotation=45) 
    plt.tight_layout()
```
